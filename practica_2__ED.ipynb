# Clase Direccion



```python
class Direccion:
    def __init__(self, *direccion):
        self._direccion = direccion

    def direccion(self):
        return {
          'calle':self._direccion[0],
          'no_calle':self._direccion[1],
          'nomenclatura':self._direccion[2],
          'barrio':self._direccion[3],
          'ciudad':self._direccion[4]
        }

    @property
    def calle(self):
      return self._direccion[0]

    @property
    def no_calle(self):
      return self._direccion[1]

    @property
    def nomenclatura(self):
      return self._direccion[2]

    @property
    def barrio(self):
      return self._direccion[3]

    @property
    def ciudad(self):
      return self._direccion[4]

    def to_string(self):
      print(self._direccion)
```

# Clase Fecha

```python
class Fecha:
    def __init__(self, dia, mes, año):
        self._dia = dia
        self._mes = mes
        self._año = año

    def fecha(self):
        return {
            'dia':self._dia,
            'mes':self._mes,
            'año':self._año
        }

    @property
    def dia(self):
      return self._dia

    @property
    def mes(self):
      return self._mes

    @property
    def año(self):
      return self._año

    @dia.setter
    def dia(self, dia):
      self._dia = dia

    @mes.setter
    def mes(self, mes):
      self._mes = mes

    @año.setter
    def año(self, año):
      self._año = año

    def to_string(self):
      print(f'{self._dia}/{self._mes}/{self._año}')
```

# Clase Usuario

```python
class Usuario:
    def __init__(self, id, nombre, fecha_nac, ciudad_nac, dir, tel, email):
        self._id = id
        self._nombre = nombre
        self._fecha_nac = Fecha(*fecha_nac)  # Llama a la clase Fecha con los argumentos de fecha_nac
        self._ciudad_nac = ciudad_nac
        self._dir = Direccion(*dir)  # Llama a la clase Direccion con el argumento direccion
        self._tel = tel
        self._email = email

    def usuario(self):
        return {
            'id':self._id,
            'nombre':self._nombre,
            'fecha_nac':self._fecha_nac.fecha(),
            'ciudad_nac':self._ciudad_nac,
            'direccion':self._dir.direccion(),
            'telefono':self._tel,
            'email':self._email
        }

    @property
    def id(self):
      return self._id

    @property
    def nombre(self):
      return self._nombre

    @property
    def fecha_nac(self):
      return self._fecha_nac

    @property
    def ciudad_nac(self):
      return self._ciudad_nac

    @property
    def direccion(self):
      return self._dir

    @property
    def telefono(self):
      return self._tel

    @property
    def email(self):
      return self._email

    @id.setter
    def id(self, id):
      self._id = id

    @nombre.setter
    def nombre(self, nombre):
      self._nombre = nombre

    @fecha_nac.setter
    def fecha_nac(self, fecha_nac):
      self._fecha_nac = fecha_nac.fecha()

    @ciudad_nac.setter
    def ciudad_nac(self, ciudad_nac):
      self._ciudad_nac = ciudad_nac

    @direccion.setter
    def direccion(self, direccion):
      self._dir = dir.direccion()

    @telefono.setter
    def telefono(self, telefono):
      self._telefono = telefono

    @email.setter
    def email(self, email):
      self._email = email

    def to_string(self):
      print(f'Usuario: Id: {self._id}, Nombre: {self._nombre}, Fecha de nacimiento: {self._fecha_nac.fecha()}, Ciudad de nacimiento: {self._ciudad_nac}, Direccion: {self._dir.direccion()}, Telefono: {self._tel}, Email: {self._email}')
```

```python
class Registro:
    def __init__(self, archivo='Empleados.txt'):
        self._archivo = archivo
        self._usuarios = []

    def cargar_usuarios(self):
        try:
            with open(self._archivo, 'r') as file:
                for line in file:
                    datos = line.strip().split(',')
                    if len(datos) == 8:
                        id, nombre, dia, mes, año, ciudad_nac, calle, no_calle, nomenclatura, barrio, ciudad, tel, email = datos
                        fecha_nac = (int(dia), int(mes), int(año))
                        direccion = (calle, no_calle, nomenclatura, barrio, ciudad)
                        usuario = Usuario(id, nombre, fecha_nac, ciudad_nac, direccion, tel, email)
                        self._usuarios.append(usuario)
        except FileNotFoundError:
            print(f"El archivo '{self._archivo}' no existe. Se creará uno nuevo.")

    def guardar_usuarios(self):
        with open(self._archivo, 'w') as file:
            for usuario in self._usuarios:
                datos = [
                    usuario.id,
                    usuario.nombre,
                    usuario.fecha_nac.dia,
                    usuario.fecha_nac.mes,
                    usuario.fecha_nac.año,
                    usuario.ciudad_nac,
                    usuario.direccion.calle,
                    usuario.direccion.no_calle,
                    usuario.direccion.nomenclatura,
                    usuario.direccion.barrio,
                    usuario.direccion.ciudad,
                    usuario.telefono,
                    usuario.email
                ]
                linea = ','.join(map(str, datos))
                file.write(linea + '\n')

    def agregar_usuario(self, usuario):
        self._usuarios.append(usuario)
        self.guardar_usuarios()

    def buscar_usuario(self, id_usuario):
        for usuario in self._usuarios:
            if usuario.id == id_usuario:
                return usuario
        return None

    def eliminar_usuario(self, id_usuario):
        usuario = self.buscar_usuario(id_usuario)
        if usuario:
            self._usuarios.remove(usuario)
            self.guardar_usuarios()
            print(f"Usuario con ID {id_usuario} eliminado.")
        else:
            print(f"Usuario con ID {id_usuario} no encontrado.")
        print()

if __name__ == "__main__":
    registro = Registro()

    # Cargar usuarios desde el archivo
    registro.cargar_usuarios()

    # Agregar un nuevo usuario
    nuevo_usuario_1 = Usuario("10", "laura", (5, 7, 2003), "medellin", ("Nueva Calle", "123", "Nomenclatura", "Nuevo Barrio", "Nueva Ciudad"), "123456789", "nuevo@correo.com")
    nuevo_usuario_2 = Usuario("2", "alex", (5, 7, 2003), "bucaramanga", ("Nueva Calle", "123", "Nomenclatura", "Nuevo Barrio", "Nueva Ciudad"), "123456789", "nuevo@correo.com")
    nuevo_usuario_3 = Usuario("3", "marco", (5, 7, 2003), "medellin", ("Nueva Calle", "123", "Nomenclatura", "Nuevo Barrio", "Nueva Ciudad"), "123456789", "nuevo@correo.com")
    nuevo_usuario_4 = Usuario("4", "kevin", (5, 7, 2003), "medellin", ("Nueva Calle", "123", "Nomenclatura", "Nuevo Barrio", "Nueva Ciudad"), "123456789", "nuevo@correo.com")
    nuevo_usuario_5 = Usuario("5", "carlos", (5, 7, 2003), "medellin", ("Nueva Calle", "123", "Nomenclatura", "Nuevo Barrio", "Nueva Ciudad"), "123456789", "nuevo@correo.com")
    nuevo_usuario_6 = Usuario("6", "edinson", (5, 7, 2003), "medellin", ("Nueva Calle", "123", "Nomenclatura", "Nuevo Barrio", "Nueva Ciudad"), "123456789", "nuevo@correo.com")
    nuevo_usuario_7 = Usuario("7", "daniel", (5, 7, 2003), "medellin", ("Nueva Calle", "123", "Nomenclatura", "Nuevo Barrio", "Nueva Ciudad"), "123456789", "nuevo@correo.com")
    nuevo_usuario_8 = Usuario("1", "daniel", (5, 7, 2003), "medellin", ("Nueva Calle", "123", "Nomenclatura", "Nuevo Barrio", "Nueva Ciudad"), "123456789", "nuevo@correo.com")


    registro.agregar_usuario(nuevo_usuario_1)
    registro.agregar_usuario(nuevo_usuario_2)
    registro.agregar_usuario(nuevo_usuario_3)
    registro.agregar_usuario(nuevo_usuario_4)
    registro.agregar_usuario(nuevo_usuario_5)
    registro.agregar_usuario(nuevo_usuario_6)
    registro.agregar_usuario(nuevo_usuario_7)
    registro.agregar_usuario(nuevo_usuario_8)

    # Buscar un usuario existente y uno no existente
    usuario_existente = registro.buscar_usuario("5")
    usuario_no_existente = registro.buscar_usuario("10")

    if usuario_existente:
        print("Usuario existente encontrado:", usuario_existente.nombre)
    else:
        print("Usuario existente no encontrado.")
    print()

    if usuario_no_existente:
        print("Usuario no existente encontrado:", usuario_no_existente.nombre)
    else:
        print("Usuario no existente no encontrado.")
    print()
    # Eliminar un usuario del registro
    registro.eliminar_usuario('1')
    registro.eliminar_usuario('1')

    # Imprimir la lista de usuarios en el registro
    print("Usuarios en el registro:")
    for usuario in registro._usuarios:
        usuario.to_string()
```

# Lista simple - Lista enlazada

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def size(self):
        return self.size

    def isEmpty(self):
        return self.size == 0

    def setSize(self, s):
        self.size = s

    def first(self):
        if self.isEmpty():
            return None
        return self.head.data

    def Last(self):
        if self.isEmpty():
            return None
        return self.tail.data

    def addFirst(self, e):
        new_node = Node(e)
        if self.isEmpty():
            self.head = self.tail = new_node
        else:
            new_node.next = self.head
            self.head = new_node
        self.size += 1

    def addLast(self, e):
        new_node = Node(e)
        if self.isEmpty():
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.size += 1

    def removeFirst(self):
        if self.isEmpty():
            return None
        data = self.head.data
        if self.size == 1:
            self.head = self.tail = None
        else:
            self.head = self.head.next
        self.size -= 1
        return data

    def removeLast(self):
        if self.isEmpty():
            return None
        data = self.tail.data
        if self.size == 1:
            self.head = self.tail = None
        else:
            current = self.head
            while current.next != self.tail:
                current = current.next
            current.next = None
            self.tail = current
        self.size -= 1
        return data

lista = LinkedList()
lista.addFirst(1)
lista.addLast(2)
lista.addLast(3)
lista.addLast(4)

print("Tamaño de la lista:", lista.size)
print("Primer elemento:", lista.first())
print("Último elemento:", lista.Last())

print("Removiendo primer elemento:", lista.removeFirst())
print("Tamaño de la lista después de eliminar el primer elemento:", lista.size)

print("Removiendo último elemento:", lista.removeLast())
print("Tamaño de la lista después de eliminar el último elemento:", lista.size)
```

# Lista doble - Lista doblemente enlazada

```python
class DoubleNode:
  def __init__(self, data):
      self.data = data
      self.next = None
      self.prev = None

class DoubleList:
  def __init__(self):
      self.head = None
      self.tail = None
      self.size = 0

  def size(self):
      return self.size

  def isEmpty(self):
      return self.size == 0

  def first(self):
      if self.isEmpty():
          return None
      return self.head

  def last(self):
      if self.isEmpty():
          return None
      return self.tail

  def addFirst(self, data):
      new_node = DoubleNode(data)
      if self.isEmpty():
          self.head = new_node
          self.tail = new_node
      else:
          new_node.next = self.head
          self.head.prev = new_node
          self.head = new_node
      self.size += 1

  def addLast(self, data):
      new_node = DoubleNode(data)
      if self.isEmpty():
          self.head = new_node
          self.tail = new_node
      else:
          new_node.prev = self.tail
          self.tail.next = new_node
          self.tail = new_node
      self.size += 1

  def removeFirst(self):
      if self.isEmpty():
          return None
      data = self.head.data
      if self.size == 1:
          self.head = None
          self.tail = None
      else:
          self.head = self.head.next
          self.head.prev = None
      self.size -= 1
      return data

  def removeLast(self):
      if self.isEmpty():
          return None
      data = self.tail.data
      if self.size == 1:
          self.head = None
          self.tail = None
      else:
          self.tail = self.tail.prev
          self.tail.next = None
      self.size -= 1
      return data

  def remove(self, node):
      if node is None:
          return None
      data = node.data
      if node.prev:
          node.prev.next = node.next
      else:
          self.head = node.next
      if node.next:
          node.next.prev = node.prev
      else:
          self.tail = node.prev
      self.size -= 1
      return data

  def addBefore(self, node, data):
      if node is None:
          return
      new_node = DoubleNode(data)
      new_node.prev = node.prev
      new_node.next = node
      if node.prev:
          node.prev.next = new_node
      else:
          self.head = new_node
      node.prev = new_node
      self.size += 1

  def addAfter(self, node, data):
      if node is None:
          return
      new_node = DoubleNode(data)
      new_node.prev = node
      new_node.next = node.next
      if node.next:
          node.next.prev = new_node
      else:
          self.tail = new_node
      node.next = new_node
      self.size += 1
# Crear una instancia de DoubleList
my_list = DoubleList()

# Agregar elementos al principio de la lista
my_list.addFirst(1)
my_list.addFirst(2)
my_list.addFirst(3)

# Imprimir el tamaño de la lista
print("Tamaño de la lista:", my_list.size)

# Imprimir el primer elemento de la lista
print("Primer elemento:", my_list.first().data)

# Imprimir el último elemento de la lista
print("Último elemento:", my_list.last().data)

# Eliminar el primer elemento de la lista y guardarlo en una variable
removed_element = my_list.removeFirst()

# Imprimir el elemento eliminado
print("Elemento eliminado:", removed_element)

# Imprimir el tamaño de la lista después de la eliminación
print("Tamaño de la lista después de eliminar:", my_list.size)

# Agregar un elemento al final de la lista
my_list.addLast(4)

# Imprimir el último elemento de la lista después de la adición
print("Último elemento después de agregar:", my_list.last().data)

# Crear un nodo y agregarlo después del primer nodo en la lista
node_to_add_after = my_list.first()
my_list.addAfter(node_to_add_after, 5)

# Imprimir el tamaño de la lista después de la adición
print("Tamaño de la lista después de agregar después:", my_list.size)

# Agregando antes y después de un nodo especifico.
my_list.addBefore(my_list.first(), 6)
my_list.addAfter(my_list.first(), 10)

# Imprimir los elementos de la lista
current_node = my_list.first()
while current_node:
  print(current_node.data)
  current_node = current_node.next
```

# Pilas

```python
class ArrayStack:
  def __init__(self, capacity):
      self.data = [None] * capacity
      self.top = -1

  def size(self):
      return self.top + 1

  def isEmpty(self):
      return self.size() == 0

  def push(self, e):
      if self.size() < len(self.data):
          self.top += 1
          self.data[self.top] = e
      else:
          print("La pila está llena. No se puede realizar la operación push.")

  def pop(self):
      if not self.isEmpty():
          popped_element = self.data[self.top]
          self.data[self.top] = None
          self.top -= 1
          return popped_element
      else:
          print("La pila está vacía. No se puede realizar la operación pop.")
          return None

  def peek(self):
      if not self.isEmpty():
          return self.data[self.top]
      else:
          print("La pila está vacía. No hay elemento en la cima.")
          return None


mi_pila = ArrayStack(5)  # Crea una pila con capacidad para 5 elementos
mi_pila.push(1)
mi_pila.push(2)
print(mi_pila.pop())  # Desapila y muestra el elemento
print(mi_pila.peek())  # Muestra el elemento en la cima
print(mi_pila.isEmpty())  # Verifica si la pila está vacía
print(mi_pila.size())  # Muestra el tamaño de la pila
```

# Colas

```python
class ArrayQueue:
    def __init__(self, capacity):
        self.data = [None] * capacity
        self.capacity = capacity
        self.front = 0
        self.rear = -1
        self.size = 0

    def is_empty(self):
        return self.size == 0

    def is_full(self):
        return self.size == self.capacity

    def enqueue(self, e):
        if self.is_full():
            raise Exception("La cola está llena")
        self.rear = (self.rear + 1) % self.capacity
        self.data[self.rear] = e
        self.size += 1

    def dequeue(self):
        if self.is_empty():
            raise Exception("La cola está vacía")
        removed_element = self.data[self.front]
        self.data[self.front] = None
        self.front = (self.front + 1) % self.capacity
        self.size -= 1
        return removed_element

    def get_size(self):
        return self.size

# Ejemplo de uso:
queue = ArrayQueue(5)
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.enqueue(4)
print("Tamaño de la cola:", queue.get_size())  # Salida: 3
print("¿La cola está vacía?", queue.is_empty())
print("¿La cola está llena?", queue.is_full())
print("Elemento eliminado de la cola:", queue.dequeue())  # Salida: 1
print("Elemento eliminado de la cola:", queue.dequeue())  # Salida: 2
```
